```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')

imagen = Image.open("../imagenes/cocodrilo,coco,descarga (8).jpg")

plt.imshow(imagen)
```

```python
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```python
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
print(imagen_mat)

```

```python
imagen_mat.shape = (imagen_gr.size[1],imagen_gr.size[0])
print(imagen_mat)
```

```python
print(imagen_mat.shape)
```

```python
plt.imshow (imagen_mat, cmap="brg")
plt.show()
```

```python
imagen_mat2 = imagen_mat / 10
print(imagen_mat2)
```

```python
imagen_mat3 = imagen_mat / 3
print(imagen_mat3)
```

```python
plt.imshow(imagen_mat2)
plt.axis('off')
plt.show()
```

```python
plt.imshow(imagen_mat3)
plt.show()
```

```python
print(np.max(imagen_mat2))
print(np.max(imagen_mat))
```

```python
print(np.min(imagen_mat2))
print(np.min(imagen_mat))
```
