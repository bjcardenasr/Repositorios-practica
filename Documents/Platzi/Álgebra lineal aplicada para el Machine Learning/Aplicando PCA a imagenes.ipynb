```python
# Crear datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura con 2 subtramas en una fila y 2 columnas
plt.subplot(1, 2, 1)  # Primera subtrama
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)  # Segunda subtrama
plt.plot(x, y2)
plt.title('Coseno')

# Mostrar la figura
plt.tight_layout()  # Opcional para ajustar los márgenes
plt.show()
```
