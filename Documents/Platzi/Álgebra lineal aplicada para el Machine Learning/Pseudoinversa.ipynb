```python
x = np.linspace (-5,5,10000) 

y_1 = -4*x + 3 
y_2 = 2*x + 5
y_3 = -3*x + 1

plt.plot (x, y_1)
plt.plot (x, y_2)
plt.plot (x, y_3)

plt.xlim (-2, 2.5)
plt.ylim (-6, 6.5)

plt.show()
```

```python
import numpy as np

# Crear una matriz
A = np.array([[1, 2], [3, 4]])

# Calcular la pseudoinversa de la matriz A
pinv_A = np.linalg.pinv(A)

print("Matriz A:")
print(A)
print("Pseudoinversa de A:")
print(pinv_A)
```

```python
import numpy as np

# Definir una matriz
A = np.array([[1, 2], [3, 4], [5, 6]])

# Calcular la pseudoinversa
pseudoinversa_A = np.linalg.pinv(A)

print("Matriz original A:")
print(A)

print("Pseudoinversa de A:")
print(pseudoinversa_A)
```
