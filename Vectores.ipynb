```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1= np.array([0,0,0,3])
v2= np.array([0,0,3,3])
```

```python
plt.xlim(-2,6)
plt.ylim(-2,6)

plt.quiver([v1[0], v2[0]],
           [v1[1], v2[1]],
           [v1[2], v2[2]],
           [v1[3], v2[3]],
           angles = 'xy', scale_units = 'xy', scale = 1,
           color = sns.color_palette()
          )

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos de ejemplo
x = np.arange(-2, 2, 0.5)
y = np.arange(-2, 2, 0.5)
u = np.ones_like(x)  # Componente x constante (1) para todos los vectores
v = np.zeros_like(y)  # Componente y constante (0) para todos los vectores

# Crear el gráfico de vectores
plt.quiver(x, y, u, v, scale=1, color='blue')

# Configurar ejes y mostrar el gráfico
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Vectores')
plt.grid()
plt.show()
```
